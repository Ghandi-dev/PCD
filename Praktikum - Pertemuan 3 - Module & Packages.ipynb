## Modules & Pakckages di dalam Python

### Apa itu Modules?
Pada saat kita membuat kode program, kemudian kita mengerjakannya ke dalam 1 file python, sehingga ada banyak baris kode yang kita tuliskan ke dalam 1 file python (*.py) tersebut. Kesulitan yang dapat terjadi adalah jika kita ingin me-maintenance kode program yang sudah kita buat jika kita menuliskan kode python pada 1 file yang besar. Kemungkinan untuk proses tracing error yang menyulitkan dan harus mengecek baris perbaris dari kode tersebut.Jika banyak developer yang mengerjakan, juga akan sulit dengan hanya 1 file tersebut. <br>
Salah satu cara untuk mengetasi hal ini adalah, kita dapat mendekomposisi dari task besar menjadi task-task kecil pada file python yang berbeda yang memungkinkan dapat dikerjakan oleh banyak developer, tidak hanya 1 developer saja. task-task kecil tersebut dapat diartikan sebagai module dalam konsep bahasa pemrograman python.<br>

jadi apa itu module? module didefinisikan sebagai file-file berisikan python definitions dan statements, yang kemudian nanti dapat diimport pada saat dibutuhkan.

Untuk menggunakan module dapat dilakukan dengan 2 cara, pertama (yang paling umum)kita menggunakan module yang sudah ada (ditulis oleh orang lain) saat mengerjakan project yang kompleks. Hal ini disebut module's user;<br>
Yang kedua kita membuat module baru, untuk kepentingan sendiri, atau membantu programmer lain sehngga dapat digunakan untuk mengerjakan projetnya, ini disebut module's supplier.
Untuk menggunakan module, kita harus mengetahui namanya terlebih dahulu. beberapa module sudah terintegrasi dengan python, bersama dengan built-in function pyhton (dari python standard library). Setiap module berisikan entities, dimana dapat berupa function, nilai konstans, classes, dan object.

Contoh module yang sering digunakan yaitu module dengan nama <mark> math </mark>. Module tersebut berisikan banyak koleksi fungsi matematika. Programmer yang membutuhkan fungsi-fungsi matematika dalam kode programnya seperti <mark>sin()</mark>, <mark>cos()</mark>, tidak harus membuat fungsi matematika tersebut, namun dapat menggunakan koleksi fungsi dari module <mark> math </mark> tersebut.

## Numpy
Numpy adalah salah satu package yang banyak digunakan. Numpy merupakan operasi numerik pada python yang dirancang untuk komputasi yang lebih cepat saat bekerja pada arrays berdimensi-n. Sifat dari array numpy adalah mutable.

> import numpy

> import numpy as np

```python
# import numpy

import numpy as np
dir(np)
```

```python
# Buat array menggunakan numpy (1 Dimensi)
list_a = [0,1,2,3,4,5]
a = np.array([0,1,2,3,4,5])
a1 = np.arange(5)
print(list_a)
print()
print(a)
print()
print(a1)
print()
print(type(a))
print(a.dtype)
print(a.shape)
print(a.ndim)
```

```python
# Akses array 1 Dimensi
print(a[0])
print(a[-1])
print(a[:4])
print(a[3:])
```

```python
# Buat array menggunakan numpy (2 Dimensi)
b = np.array([[4,5.,8,4],[6,.3,2,1],[8.,6,4,.3]])
print(b)
print(type(b))
print(b.dtype)
print(b.shape)
print(b.ndim)
```

```python
# Akses array 2 Dimensi
print(b[0])
print(b[0][1])
print(b[1][1:4])
print(b[0, 1:4])
```

```python
# Buat array menggunakan numpy (3 Dimensi)
c = np.array([[[4,5,8,4],[6,3,2,1]],[[8,6,4,3], [4,7,8,9]]])
print(c)
print()
print(type(c))
print(c.dtype)
print(c.shape)
print(c.ndim)
```

```python
# Akses array 3 Dimensi
print(c[0])
print()
print(c[0][1])
print()
print(c[1][1][1:3])
print()
print(c[0, 1, 2:])
```

```python
# membuat Array dengan nilai 0 (zeros), 1 (ones), empty
_a = np.zeros(5)
_b = np.ones((4,3))
_c = np.empty((2,2,4), dtype=np.unicode_)
print(_a)
print()
print(_b)
print()
print(_c)
```

```python
# Mengisi elemen pada numpy
_b[0][1] = 4
_b[1][2] = 6
print(_b)
```

```python
# Operasi pada array
print(_b + _b)
print()
print(_b * _b)
print()
print(1/_b)
```

#### Beberapa fungsi dan method pada numpy

```python
a = np.arange(6)
print(np.sqrt(a))
print(np.max(a))
print(np.min(a))
print(np.exp(a))
print(a.sum())
print(a.mean())
print(np.mean(a))
print(np.sort(a))
```

```python
arr = np.random.randn(10)
print(arr)
print()
print(np.where(arr> 0, True, False))
print()
print(np.where(arr> 0, True, arr))
print()
print(np.where(arr> 0, arr, False))
```

```python
# Unique and other set logic
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe', 'Thomas'])
print(np.unique(names))
print(np.in1d(names, 'Will'))
```

## Pandas
Pandas merupakan package yang banyak dipakai dalam analisis numerik di dalam python yang bekerja dengan bentuk <mark>**series**</mark> (kolom), dan juga <mark>**dataframe**</mark>(kolom multi-dimensi)
<br>
dengan Pandas, kita dapat membuat representasi visualisasi yang lebih baik dari multi-dimensi data seperti <mark>dictionary</mark>, menjadi data tabular yang nilai keys dinyatakan sebagai header kolom, sedangkan values adalah komponen/elemen kolom.

```python
#!pip install pandas
```

```python
import pandas as pd

# buat dictionary
dct = {'aple': np.array([1,2,3,4]),
       'oranges' : np.array([6,7,8,9])
      }
print(dct)
```

```python
df = pd.DataFrame(dct)
df.index = ['A', 'B', 'C', 'D']
print(df)
```

```python
# Akses elemen data frame

# by row
print(df.loc['A'])
print(df.loc['C']['aple'])
print(df.loc['C']['oranges'])
print()

#by column
print(df['aple'])
print(df['oranges'])
print()
print(df['oranges']['A'])
print(df.oranges['A'])
```
