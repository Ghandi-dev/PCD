```python
import cv2
# from google.colab.patches import cv2_imshow
import numpy as np

# buka citra
img = cv2.imread('daun.jpg')
```

```python
cv2.imshow('display',img)
```

```python
row, col, ch = img.shape

kanvas = np.zeros((row, col, 3), np.uint8)
```

```python
kanvas
```

```python
for i in range (0, row):
  for j in range (0, col):
    if (img[i,j,0] < 150 and img[i,j,1] < 150 and img[i,j,2] < 150):
      kanvas[i,j,0] = 255
      kanvas[i,j,1] = 255
      kanvas[i,j,2] = 255
    else:
      kanvas[i,j] = img[i,j]

cv2.imshow("2",kanvas)
```
